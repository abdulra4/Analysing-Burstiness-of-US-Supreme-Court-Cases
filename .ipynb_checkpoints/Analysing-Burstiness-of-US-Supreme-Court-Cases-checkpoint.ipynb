## 2) Getting Revisions & Bustiness from MediaWiki

```python
# Get revisions to calculate Burstiness
article_data = {}

def getRevisions(title, rv_prop='timestamp|user', 
                 rv_start='today', rv_end='2000-01-01T00:00:00Z'):
    title = title
    
    # My parameters from MediaWiki
    action = 'query'
    prop = 'revisions'
    rv_limit = 'max'
    today = dt.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%dT%H:%M:%SZ')
    user = 'Chrome (Macbook Pro)'
    
    # Managing errors for start date
    if rv_start == "today":
        rv_start = today
    else:
        rv_start = start_date
    
    # Temporary dictionary holding amount of revisions for each user
    revisions_by_user = {}
    
    # Connect to Wikipedia
    session = mw.Session('https://en.wikipedia.org', user_agent=user)

    # Query Wikipedia for revisions on the supplied article
    # The result is stored into the dictionary "rev_dict"
    rev_dict = session.get(action=action,
                           prop=prop,
                           rvprop=rv_prop,
                           titles=title,
                           rvlimit=rv_limit,
                           rvstart=rv_start,
                           rvend=rv_end)
    
    # Find page_id for selected article
    for keys in rev_dict['query']['pages'].keys():
        page_id = keys
        
    # Go through the timestamps for each revision made.
    # If the timestamp is already a key in our dictionary
    rev_timestamps = []
    for props in rev_dict['query']['pages'][str(page_id)]['revisions']:
        if 'user' in props:
            if (props['user'] not in revisions_by_user):
                revisions_by_user[props['user']] = 1
            else:
                revisions_by_user[props['user']] += 1

        timestamp = dup.parse(props['timestamp'])
        rev_timestamps.append(timestamp)
        
    # Check if there is a section named "continue".
    while 'continue' in rev_dict:
        continue_val = rev_dict['continue']['rvcontinue']
        rev_dict = session.get(action=action,
                               prop=prop,
                               rvprop=rv_prop,
                               titles=title,
                               rvlimit=rv_limit,
                               rvstart=today,
                               rvend=rv_end,
                               rvcontinue=continue_val)
        for props in rev_dict['query']['pages'][str(page_id)]['revisions']:
            if 'user' in props:
                if (props['user'] not in revisions_by_user):
                    revisions_by_user[props['user']] = 1
                else:
                    revisions_by_user[props['user']] += 1
            timestamp = dup.parse(props['timestamp'])
            rev_timestamps.append(timestamp)

    # List of tuples of revisions made by user for page
    sorted_user_revisions = sorted(revisions_by_user.items(),
                                   key=operator.itemgetter(1))[::-1]
    
    # Enumerate the times between events into a list
    interevent_times = []
    for i in range(0, len(rev_timestamps)-1):
        interevent_times.append((rev_timestamps[i]\ 
                                 - rev_timestamps[i+1]).total_seconds())
    
    # Add data to global dictionaries
    article_data[title] = {}
    article_data[title]['user_revs'] = sorted_user_revisions
    article_data[title]['revision_times'] = rev_timestamps
    article_data[title]['interevent_times'] = interevent_times
    getBurstiness(article)
    
def getBurstiness(title):
    # Generate probability distribution for data
    hist, bins_e = np.histogram(article_data[title]['interevent_times'], 
                                bins=100, density=True)
    # Calculate interevent time mean and standard deviation
    interevent_mean = (np.mean(hist))
    interevent_std_dev = (np.std(hist))
    Burstiness = ((interevent_std_dev - interevent_mean) \
                  / (interevent_std_dev + interevent_mean))
    article_data[title]['Burstiness'] = Burstiness
```

## 3) Data Wrangling: Merging with SCDB

```python
Burstiness = []

for article in uscw['Page title']:
    getRevisions(article)
    Burstiness.append(article_data[article]['Burstiness'])

ind = np.arange(len(article_data.keys()))
article_editors = []
article_edits = []

article_count = 0
for article in article_data.keys():
    article_editors.append(0)
    article_edits.append(0)
    for tuple in article_data[article]['user_revs']:
        article_editors[article_count] += 1
        article_edits[article_count] += tuple[1]
    article_count += 1
    
uscw['TotalEdits'] = article_edits
uscw['TotalEditors'] = article_editors
uscw['Burstiness'] = Burstiness
```

```python
cite = uscw.Content.str.findall(r'\d+ U.S. [0-9]+')
citation = []

for i in cite:
    try:
        citation.append(i[0])
        
    except:
        citation.append(np.nan)
        continue

uscw['usCite'] = citation

# Merging with US Supreme Court Database
uscc = pd.read_excel('SCDB_2018_02_caseCentered_Citation.xlsx')
uscw = uscw.dropna()
uscmerged = uscw.merge(uscc, how='inner', on='usCite')
uscmerged.columns = ['rank', 'pageTitle', 'pageViews', 'dailyAverage', 
                     'assessment', 'importance', 'content', 'totalEdits', 
                     'totalEditors', 'burstiness', 'usCite', 'caseName',
                     'caseId', 'dateDecision', 'decisionType', 'naturalCourt',
                     'chief', 'petitioner', 'respondent', 'jurisdiction',
                     'caseOriginState', 'caseDisposition', 'partyWinning', 'issueArea',
                     'majVotes', 'minVotes']
# Save
uscmerged.to_csv('uscmerged_data.csv')
```
